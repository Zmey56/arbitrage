```python
import json
import requests
import pandas as pd
```

```python
headers = {
    "Accept": "*/*",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8",
    "Cache-Control": "no-cache",
    "Connection": "keep-alive",
    "Content-Length": "123",
    "content-type": "application/json",
    "Host": "p2p.binance.com",
    "Origin": "https://p2p.binance.com",
    "Pragma": "no-cache",
    "TE": "Trailers",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:88.0) Gecko/20100101 Firefox/88.0"
}

data = {
  "asset": "BTC",
  "fiat": "RUB",
  "merchantCheck": False,
  "page": 3,
  "payTypes": [],
  "publisherType": None,
  "rows": 10,
  "tradeType": "Buy"
}


r = requests.post('https://p2p.binance.com/bapi/c2c/v2/friendly/c2c/adv/search', headers=headers, json=data)
```

```python
r
```

```python
data = r.json()
```

```python
data
```

```python
df = pd.DataFrame(d["adv"] for d in data["data"])
```

```python
df = df.drop(['classify', 
              'advStatus', 
              'priceType', 
              'priceFloatingRatio', 
              'rateFloatingRatio', 
              'currencyRate',
             'tradeMethodCommissionRates',
              'launchCountry', 
              'abnormalStatusList',
              'closeReason', 
              'initAmount', 
              'amountAfterEditing'], axis=1)
```
